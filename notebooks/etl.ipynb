```python
import pandas as pd
from glob import  glob

pd.set_option('display.max_columns', None)
```

```python
arquivo  = glob(r"../../2. Dados Recebidos/Base Campinas/Custo*_v2.csv")[0]
```

```python
arquivo.split('_')[-2]
```

```python
df = pd.DataFrame()
colunas = pd.read_csv(glob(r"../../2. Dados Recebidos/Base Campinas/Custo*_v2.csv")[0]).columns

for arquivo in glob(r"../../2. Dados Recebidos/Base Campinas/Custo*_v2.csv"):
    df_temp = pd.read_csv(arquivo, names=colunas, skiprows=1)
    df_temp['logomarca']= arquivo.split('_')[-2]
    df = pd.concat([df, df_temp], ignore_index=True)
```

```python
df.to_csv('../../2. Dados Recebidos/Base Campinas/Custo_Campinas_Completo_v2.csv', index=False)
```
